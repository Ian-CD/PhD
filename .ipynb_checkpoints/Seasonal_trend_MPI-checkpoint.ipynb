```python
import xarray as xr
import pandas as pd
import numpy as np
import calendar as cld
import matplotlib.pyplot as plt
import matplotlib.colors
colors_land = plt.cm.terrain(np.linspace(0.25, 1, 256))
import proplot as pplt # New plot library (https://proplot.readthedocs.io/en/latest/)
pplt.rc['savefig.dpi'] = 300 # 1200 is too big! #https://proplot.readthedocs.io/en/latest/basics.html#Creating-figures
from scipy.stats import chi2
from numba import njit,prange
import matplotlib as mpl
mpl.rcParams['hatch.linewidth'] = 0.05  # previous pdf hatch linewidth
from matplotlib.dates import DateFormatter

from scipy.stats import linregress
```

```python
normal_format = {'gridlinewidth':0.1, 'gridcolor':'gray8', 'gridalpha':0.5, 'coast':True,'borders':True ,'reso':'hi', 'labels':True, 'lonlines':2, 'latlines':2, 'abc':False, 'latlim':[43.,48.5],'lonlim':[4.,16.]}
multiplot_format = {'gridlinewidth':0.1, 'gridcolor':'gray8', 'gridalpha':0.5, 'coast':True,'borders':True ,'reso':'hi', 'labels':False, 'lonlines':2, 'latlines':2, 'abc':False, 'latlim':[43.,48.5],'lonlim':[4.,16.]}

imin = 32 ; imax = -30
jmin = 20 ; jmax = -15
ds = xr.open_dataset('/bettik/beaumetj/MARout/MAR-ERA-20C/MARgrid_EUf.nc')
lon = ds.LON[jmin:jmax,imin:imax]
lat = ds.LAT[jmin:jmax,imin:imax]
H = np.array(ds.SH[jmin:jmax,imin:imax])
```

```python
wp_meanseason_meanT_245 = np.load('/bettik/castelli/data/saved_data_MAR_MPI_ESM1_2_HR/SSP245/wp_meanseason_meanT.npy')[:,:,jmin:jmax,imin:imax]
origin_T_245=np.full(wp_meanseason_meanT_245.shape[1:4],np.nan)
slope_T_245=np.full(wp_meanseason_meanT_245.shape[1:4],np.nan)
pvalue_T_245=np.full(wp_meanseason_meanT_245.shape[1:4],np.nan)
for i in range(wp_meanseason_meanT_245.shape[3]):
    # print(str(lon)+'/'+str(les_season_maxT.shape[3]),end=' ')
    for j in range(wp_meanseason_meanT_245.shape[2]):
        for season in range(4):
            linregress_T_245 = linregress(np.arange(wp_meanseason_meanT_245.shape[0]), wp_meanseason_meanT_245[:,season,j,i])
            origin_T_245[season][j][i] = linregress_T_245.intercept
            slope_T_245[season][j][i] = linregress_T_245.slope
            pvalue_T_245[season][j][i] = linregress_T_245.pvalue
```

```python
wp_meanseason_meanT_585 = np.load('/bettik/castelli/data/saved_data_MAR_MPI_ESM1_2_HR/SSP585/wp_meanseason_meanT.npy')[:,:,jmin:jmax,imin:imax]
origin_T_585=np.full(wp_meanseason_meanT_585.shape[1:4],np.nan)
slope_T_585=np.full(wp_meanseason_meanT_585.shape[1:4],np.nan)
pvalue_T_585=np.full(wp_meanseason_meanT_585.shape[1:4],np.nan)
for i in range(wp_meanseason_meanT_585.shape[3]):
    # print(str(lon)+'/'+str(les_season_maxT.shape[3]),end=' ')
    for j in range(wp_meanseason_meanT_585.shape[2]):
        for season in range(4):
            linregress_T_585 = linregress(np.arange(wp_meanseason_meanT_585.shape[0]), wp_meanseason_meanT_585[:,season,j,i])
            origin_T_585[season][j][i] = linregress_T_585.intercept
            slope_T_585[season][j][i] = linregress_T_585.slope
            pvalue_T_585[season][j][i] = linregress_T_585.pvalue
```

```python
f, axs = pplt.subplots(proj='cyl',ncols=4,nrows=2)

for i in range(4):
    ax = axs[i]
    cb = ax.pcolormesh(lon,lat,100*slope_T_245[i],levels=np.linspace(0,8,65),cmap='Reds')
    ax.text(5,47.5,"{:.1f}".format(np.mean(100*slope_T_245[i])),ha='center',va='center',fontsize=20)
    
    ax = axs[i+4]
    cb = ax.pcolormesh(lon,lat,100*slope_T_585[i],levels=np.linspace(0,8,65),cmap='Reds')
    ax.text(5,47.5,"{:.1f}".format(np.mean(100*slope_T_585[i])),ha='center',va='center',fontsize=20)

f.colorbar(cb, label= 'C /centuries',ticks=np.linspace(0,8,9))

axs.format(**multiplot_format,suptitle='Trend of mean temperature',collabels=['DJF','MAM','JJA', 'SON'],rowlabels=['SSP245','SSP585'])
```

#### Scatter plots for trends as a function of altitude

```python
wp_meanseason_meanT_hist = wp_meanseason_meanT_245[:54,:,:,:]
wp_meanseason_meanT_245 = wp_meanseason_meanT_245[55:,:,:,:]
wp_meanseason_meanT_585 = wp_meanseason_meanT_585[55:,:,:,:]

origin_T_hist=np.full(wp_meanseason_meanT_hist.shape[1:4],np.nan)
slope_T_hist=np.full(wp_meanseason_meanT_hist.shape[1:4],np.nan)
pvalue_T_hist=np.full(wp_meanseason_meanT_hist.shape[1:4],np.nan)

origin_T_245=np.full(wp_meanseason_meanT_245.shape[1:4],np.nan)
slope_T_245=np.full(wp_meanseason_meanT_245.shape[1:4],np.nan)
pvalue_T_245=np.full(wp_meanseason_meanT_245.shape[1:4],np.nan)

origin_T_585=np.full(wp_meanseason_meanT_585.shape[1:4],np.nan)
slope_T_585=np.full(wp_meanseason_meanT_585.shape[1:4],np.nan)
pvalue_T_585=np.full(wp_meanseason_meanT_585.shape[1:4],np.nan)

for i in range(wp_meanseason_meanT_585.shape[3]):
    # print(str(lon)+'/'+str(les_season_maxT.shape[3]),end=' ')
    for j in range(wp_meanseason_meanT_585.shape[2]):
        for season in range(4):
            linregress_T_hist = linregress(np.arange(wp_meanseason_meanT_hist.shape[0]), wp_meanseason_meanT_hist[:,season,j,i])
            origin_T_hist[season][j][i] = linregress_T_hist.intercept
            slope_T_hist[season][j][i] = linregress_T_hist.slope
            pvalue_T_hist[season][j][i] = linregress_T_hist.pvalue
            
            linregress_T_245 = linregress(np.arange(wp_meanseason_meanT_245.shape[0]), wp_meanseason_meanT_245[:,season,j,i])
            origin_T_245[season][j][i] = linregress_T_245.intercept
            slope_T_245[season][j][i] = linregress_T_245.slope
            pvalue_T_245[season][j][i] = linregress_T_245.pvalue
            
            linregress_T_585 = linregress(np.arange(wp_meanseason_meanT_585.shape[0]), wp_meanseason_meanT_585[:,season,j,i])
            origin_T_585[season][j][i] = linregress_T_585.intercept
            slope_T_585[season][j][i] = linregress_T_585.slope
            pvalue_T_585[season][j][i] = linregress_T_585.pvalue
```

```python
def detect_alps(H):
    nlat,nlon = np.shape(H)
    mask = np.bool8(np.zeros((nlat,nlon)))
    r = 4
    for j in range(r,nlat-r):
        for i in range(r,nlon-r):
            mask[j,i] = np.logical_and(H[j,i]>360 ,np.any(H[j-r:j+r,i-r:i+r]>1300))
            # mask[j,i] = np.std(H[j-r:j+r,i-r:i+r])>200
    return mask
alps = detect_alps(H)
alps[lon<4.8] = False
alps[np.logical_and(lon>10,lat<45.2)] = False

north_alps = np.copy(alps)
north_alps[lon>8.6] = False
north_alps[lat<45] = False
#north_alps[lat>46.5] = False

south_alps = np.copy(alps)
south_alps[lat>45] = False

east_alps = np.copy(alps)
east_alps[lon<8.6] = False
```

```python
f, axs = pplt.subplots(proj='cyl', ncols=2, nrows=2, axwidth=8)

m=axs[0].pcolormesh(lon, lat, H, levels=34,cmap=colors_land)
f.colorbar(m,label= 'm.asl')
axs[0].contour(lon, lat,alps)
axs[0].format(**normal_format,title='Alps region')
axs[0].format(ocean=True,oceancolor='royalblue',reso='hi')


m=axs[1].pcolormesh(lon, lat, H, levels=34,cmap=colors_land)
axs[1].contour(lon, lat,north_alps)
axs[1].format(**normal_format,title='Northern Alps region')
axs[1].format(ocean=True,oceancolor='royalblue',reso='hi')


m=axs[2].pcolormesh(lon, lat, H, levels=34,cmap=colors_land)
axs[2].contour(lon, lat,south_alps)
axs[2].format(**normal_format,title='Southern Alps region')
axs[2].format(ocean=True,oceancolor='royalblue',reso='hi')


m=axs[3].pcolormesh(lon, lat, H, levels=34,cmap=colors_land)
axs[3].contour(lon, lat,east_alps)
axs[3].format(**normal_format,title='Eastern Alps region')
axs[3].format(ocean=True,oceancolor='royalblue',reso='hi')
```

```python
f, axs = pplt.subplots(ncols=4, nrows=4)
leg = ['hist','SSP245','SSP585']

for i in range(4):
    slope_T_hist_alps = np.ma.masked_array(slope_T_hist[i], mask=np.invert(alps))
    slope_T_hist_east_alps = np.ma.masked_array(slope_T_hist[i], mask=np.invert(east_alps))
    slope_T_hist_north_alps = np.ma.masked_array(slope_T_hist[i], mask=np.invert(north_alps))
    slope_T_hist_south_alps = np.ma.masked_array(slope_T_hist[i], mask=np.invert(south_alps))
    slopes=[slope_T_hist_alps,slope_T_hist_east_alps,slope_T_hist_north_alps,slope_T_hist_south_alps]
    
    for j,slope in enumerate(slopes):
        ax = axs[i+4*j]
        blue_dot = ax.scatter(10*slope,ds.SH[jmin:jmax,imin:imax],s=2,alpha=0.5)
        
    
for i in range(4):
    slope_T_245_alps = np.ma.masked_array(slope_T_245[i], mask=np.invert(alps))
    slope_T_245_east_alps = np.ma.masked_array(slope_T_245[i], mask=np.invert(east_alps))
    slope_T_245_north_alps = np.ma.masked_array(slope_T_245[i], mask=np.invert(north_alps))
    slope_T_245_south_alps = np.ma.masked_array(slope_T_245[i], mask=np.invert(south_alps))
    slopes=[slope_T_245_alps,slope_T_245_east_alps,slope_T_245_north_alps,slope_T_245_south_alps]
    
    for j,slope in enumerate(slopes):
        ax = axs[i+4*j]
        orange_dot = ax.scatter(10*slope,ds.SH[jmin:jmax,imin:imax],s=2,alpha=0.5)


for i in range(4):
    slope_T_585_alps = np.ma.masked_array(slope_T_585[i], mask=np.invert(alps))
    slope_T_585_east_alps = np.ma.masked_array(slope_T_585[i], mask=np.invert(east_alps))
    slope_T_585_north_alps = np.ma.masked_array(slope_T_585[i], mask=np.invert(north_alps))
    slope_T_585_south_alps = np.ma.masked_array(slope_T_585[i], mask=np.invert(south_alps))
    slopes=[slope_T_585_alps,slope_T_585_east_alps,slope_T_585_north_alps,slope_T_585_south_alps]
    
    for j,slope in enumerate(slopes):
        ax = axs[i+4*j]
        green_dot = ax.scatter(10*slope,ds.SH[jmin:jmax,imin:imax],s=2,alpha=0.5)
        ax.set_xlim((-0.3,1))
        ax.set_ylim((0.,3500))
        if(j==3):
            ax.legend([blue_dot,orange_dot,green_dot],leg,loc='bottom')

axs.format(suptitle='Trend of mean temperature by region',collabels=['DJF','MAM','JJA', 'SON'],
           rowlabels=['Alps','Eastern Alps','Northern Alps','Southern Alps'],xlabel='Tmean trend (C/decade)',ylabel='Elevation (m.a.s.l.)')
```
